```python
# Reading the Image
image = cv2.imread('Taj-Mahal-Agra-India.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # BGR to RGB , for plotting on matplotlib
fig, ax = plt.subplots(1, figsize=(10,6)) # Making a subplot of size 10,6
plt.imshow(image) 
```

# Now let us take one example of Kernel and see what happens

```python
kernel = np.ones((3, 3), np.float32) / 9
kernel
```

```python
img = cv2.filter2D(image, -1, kernel) # Applying the filter on the Image
fig, ax = plt.subplots(1, figsize=(10,6)) # Creating the subplot of size 10,6
plt.imshow(img)
```

# Nothing Special but let us now make our kernel more sharp

```python
kernel = np.array([[0, -1, 0],
                   [-2, 6, -2],
                   [0, -1, 0]])

kernel2 = np.array([[0.0, -1.0, 0.0], 
                   [-1.0, 4.0, -1.0],
                   [0.0, -1.0, 0.0]])
```

```python
img = cv2.filter2D(image, -1, kernel)
fig, ax = plt.subplots(1, figsize=(10,6))
plt.imshow(img)
```

# And we can see the edges are clearly visible with it

```python
img = cv2.filter2D(image, -1, kernel2)
fig, ax = plt.subplots(1, figsize=(10,6))
plt.imshow(img)
```

# After applying the second sharp kernel. we can see the middle part of Image is more enhanced compare with other parts

```python
image = cv2.imread('Taj-Mahal-Agra-India.jpg')
image = cv2.blur(img, (5,5))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(1, figsize=(12,8))
plt.imshow(image)
```

```python
image = cv2.imread('Taj-Mahal-Agra-India.jpg')
# convert to black and white
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
r ,image = cv2.threshold(image, 150, 255, cv2.THRESH_BINARY)
# create kernel
kernel = np.ones((5,5), np.uint8)
fig, ax = plt.subplots(1, figsize=(16,12))
# original
ax = plt.subplot(232)
plt.imshow(image, cmap='Greys')
plt.title('original')
# erosion
e = cv2.erode(image, kernel)
ax = plt.subplot(234)
plt.imshow(e, cmap='Greys')
plt.title('erosion')
# dilation
d = cv2.dilate(image, kernel)
ax = plt.subplot(235)
plt.imshow(d, cmap='Greys')
plt.title('dilation')
# morphological gradient (dilation - erosion)
m = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)
ax = plt.subplot(236)
plt.imshow(m, cmap='Greys')
plt.title('dilation - erosion')
```

# Now let us apply these techniques with cv2 in easy manner

```python
image = cv2.imread('Taj-Mahal-Agra-India.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(1, figsize=(12,16))
ax = plt.subplot(211)
g = cv2.GaussianBlur(image, (3,3), 0)
plt.imshow(g)
ax = plt.subplot(212)
m = cv2.medianBlur(image, 3)
plt.imshow(m)
```
